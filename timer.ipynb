```python
import time
import math
import tkinter as tk

class Timer:
    def __init__(self):
        self.one = 0
        #時間の定義
        self.hour = 0 #時
        self.min = 0 #分
        self.sec = 0 #秒
        self.comma = 0 #コンマ秒

        #スタートを押した回数でスタートかストップを判断
        self.judge = 0
        
        #tkinterの設定
        self.root = tk.Tk()
        self.canvas = tk.Canvas(self.root, width=800, height=450)
        
        #使用するラベルのリスト
        self.labels = [tk.Label(self.root, text="時"),
                       tk.Label(self.root, text="分"),
                       tk.Label(self.root, text="秒"),
                       tk.Label(self.root, text="{:d} : {:0>2d} : {:0>2d}.{:0>2d}".format(math.floor(self.hour), math.floor(self.min), math.floor(self.sec), int(self.comma)), font=("normal", "40", "bold"))]
        
        #使用するボタンのリスト
        self.buttons = [tk.Button(self.root, text="スタート/ストップ", command=lambda : self.push_start(1)),
                        tk.Button(self.root, text="+", repeatdelay=500, repeatinterval=50, command=self.push_hour_plus),
                        tk.Button(self.root, text="-", repeatdelay=500, repeatinterval=50 ,command=self.push_hour_minus),
                        tk.Button(self.root, text="+", repeatdelay=500, repeatinterval=50 ,command=self.push_min_plus),
                        tk.Button(self.root, text="-", repeatdelay=500, repeatinterval=50 ,command=self.push_min_minus),
                        tk.Button(self.root, text="+", repeatdelay=500, repeatinterval=50 ,command=self.push_sec_plus),
                        tk.Button(self.root, text="-", repeatdelay=500, repeatinterval=50 ,command=self.push_sec_minus),
                        tk.Button(self.root, text="リセット", repeatdelay=500, repeatinterval=50 ,command=self.push_reset)]
        
        self.create()
        
    #初期設定
    def create(self):
        self.root.title("Timer")
        self.root.geometry("800x450")

        self.canvas.place(x=0, y=0)

        self.labels[0].pack(padx=(100, 0), pady=(100,0), side=tk.LEFT)
        self.buttons[1].pack(padx=(0,0),pady=(100,0), side=tk.LEFT)
        self.buttons[2].pack(padx=(0,0),pady=(100, 0), side=tk.LEFT)

        self.labels[1].pack(padx=(100, 0), pady=(100,0), side=tk.LEFT)
        self.buttons[3].pack(padx=(0,0),pady=(100, 0), side=tk.LEFT)
        self.buttons[4].pack(padx=(0,0),pady=(100, 0), side=tk.LEFT)

        self.labels[2].pack(padx=(100, 0), pady=(100,0), side=tk.LEFT)
        self.buttons[5].pack(padx=(0,0),pady=(100, 0), side=tk.LEFT)
        self.buttons[6].pack(padx=(0,0),pady=(100, 0), side=tk.LEFT)

        self.buttons[7].pack(padx=(0, 0), pady=(200, 0))
        self.buttons[0].pack(padx=(200, 0), pady=(0, 0))

        self.labels[3].pack(before=self.labels[0], padx=100, pady=50)

    #それぞれのボタンを押したときの関数
    def push_hour_plus(self):           
        self.hour += 1
        self.replace()

    def push_hour_minus(self):
        if self.hour == 0:
            pass
        else:
            self.hour -= 1
        self.replace()

    def push_min_plus(self):
        if self.min == 59:
            self.min = 0
            self.hour += 1
        else:
            self.min += 1
        self.replace()
    
    def push_min_minus(self):
        if self.min == 0:
            pass
        else:
            self.min -= 1
        self.replace()

    def push_sec_plus(self):
        if self.sec == 59 and self.min == 59:
            self.sec = 0
            self.min = 0
            self.hour += 1
        elif self.sec == 59:
            self.sec = 0
            self.min += 1
        else:
            self.sec += 1
        self.replace()
    
    def push_sec_minus(self):
        if self.sec == 0:
            pass
        else:
            self.sec -= 1
        self.replace()

    def push_reset(self):
        self.comma = 0
        self.sec = 0
        self.min = 0
        self.hour = 0
        self.replace()

    def push_start(self, check=0): 
    #スタートを１回押すとjudgeが１になりカウントダウンを開始する。
    #時間が０になるまではjudgeが１のままでカウントダウンし続ける
    #スタートをもう一度押すか時間が０になるとjudgeが２になりカウントダウンをやめる
        self.judge += check
        if self.judge == 1: 
            if self.one == 0:
                self.buttons[1]["state"] = tk.DISABLED
                self.buttons[2]["state"] = tk.DISABLED
                self.buttons[3]["state"] = tk.DISABLED
                self.buttons[4]["state"] = tk.DISABLED
                self.buttons[5]["state"] = tk.DISABLED
                self.buttons[6]["state"] = tk.DISABLED
                self.buttons[7]["state"] = tk.DISABLED
                self.one = 1
            #self.root.after(10, self.count())
            start = time.time()
            while time.time() - start < 0.01:
                pass
            self.count()
            
        elif self.judge == 2:
            self.judge = 0
            self.buttons[1]["state"] = tk.NORMAL
            self.buttons[2]["state"] = tk.NORMAL
            self.buttons[3]["state"] = tk.NORMAL
            self.buttons[4]["state"] = tk.NORMAL
            self.buttons[5]["state"] = tk.NORMAL
            self.buttons[6]["state"] = tk.NORMAL
            self.buttons[7]["state"] = tk.NORMAL
            self.root.mainloop()


    #時間の表示を変更
    def replace(self):
        self.labels[3]["text"] = "{:d} : {:0>2d} : {:0>2d}.{:0>2d}".format(math.floor(self.hour), math.floor(self.min), math.floor(self.sec), int(self.comma))
        
    #時間変更
    def change_sec(self):
        self.sec -= 0.01
        self.comma = (self.sec - math.floor(self.sec))*100
        self.replace()

    def change_min(self):
        self.min -= 1
        self.sec = 60
        self.replace()

    def change_hour(self):
        self.hour -= 1
        self.sec = 60
        self.min = 59
        self.replace()

    def count(self):
        if self.sec >= 0.01:
            self.change_sec()
        elif self.min > 0:
            self.change_min()
        elif self.hour > 0:
            self.change_hour()
        else:
            self.push_start(1)

        self.root.update()
        self.push_start()


#メインプログラム
timer = Timer()
timer.root.mainloop()
```
